# Limpiando data JSON de spotify

## Librerias

```python
#librerias 
from pathlib import Path
import json

import pandas as pd
import numpy as np
```

## Parsear los JSON

```python
#Definiendo los paths
RAW_SPOTIFY_PATH = Path("/Users/user27/Library/CloudStorage/OneDrive-WPPCloud/Documents/music-letterboxd-eda/data/raw/spotify")
PROCESSED_PATH = Path("/Users/user27/Library/CloudStorage/OneDrive-WPPCloud/Documents/music-letterboxd-eda/data/processed/spotify")

PROCESSED_PATH.mkdir(parents=True, exist_ok=True)
```

```python
json_files = sorted(RAW_SPOTIFY_PATH.glob("*.json"))

print(f"Total JSON files found: {len(json_files)}")
for file in json_files:
    print(file.name)
```

```python
dfs = []

for file in json_files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    df_temp = pd.DataFrame(data)
    df_temp["source_file"] = file.name
    df_temp["source_type"] = "video" if "Video" in file.name else "audio"
    
    dfs.append(df_temp)

spotify_raw = pd.concat(dfs, ignore_index=True)
```

```python
print(f"Shape: {spotify_raw.shape}")
spotify_raw.head()
```

```python
file_summary = (
    spotify_raw
    .groupby(["source_file", "source_type"])
    .size()
    .reset_index(name="rows")
    .sort_values("source_file")
)

file_summary
```

# Limpiar el dataset

```python
spotify_clean = spotify_raw.copy()
```

```python
spotify_clean["ts"] = pd.to_datetime(spotify_clean["ts"], utc=True, errors="coerce")

spotify_clean["ms_played"] = pd.to_numeric(spotify_clean["ms_played"], errors="coerce")

spotify_clean["offline_timestamp"] = pd.to_datetime(
    spotify_clean["offline_timestamp"],
    unit="ms",
    utc=True,
    errors="coerce"
)
```

```python
spotify_clean["offline"].value_counts(dropna=False)
```

```python
spotify_clean["offline"] = spotify_clean["offline"].fillna(False).astype(bool)
```

```python
spotify_clean["date"] = spotify_clean["ts"].dt.date
spotify_clean["year"] = spotify_clean["ts"].dt.year
spotify_clean["month"] = spotify_clean["ts"].dt.month
spotify_clean["day"] = spotify_clean["ts"].dt.day
spotify_clean["hour"] = spotify_clean["ts"].dt.hour
spotify_clean["weekday"] = spotify_clean["ts"].dt.day_name()

spotify_clean["minutes_played"] = spotify_clean["ms_played"] / 60000
spotify_clean["hours_played"] = spotify_clean["ms_played"] / 3600000
```

```python
spotify_clean["content_type"] = np.select(
    [
        spotify_clean["spotify_track_uri"].notna() | spotify_clean["master_metadata_track_name"].notna(),
        spotify_clean["spotify_episode_uri"].notna() | spotify_clean["episode_name"].notna(),
        spotify_clean["audiobook_uri"].notna() | spotify_clean["audiobook_title"].notna(),
        spotify_clean["source_type"].eq("video")
    ],
    [
        "track",
        "episode",
        "audiobook",
        "video"
    ],
    default="unknown"
)
```

```python
spotify_clean["is_valid_stream"] = spotify_clean["ms_played"] >= 30000
spotify_clean["is_short_stream"] = spotify_clean["ms_played"] < 30000
spotify_clean["is_very_short_stream"] = spotify_clean["ms_played"] < 10000
spotify_clean["is_skipped_custom"] = spotify_clean["skipped"] | (spotify_clean["ms_played"] < 10000)
```

```python
ordered_cols = [
    # tiempo
    "ts", "date", "year", "month", "day", "hour", "weekday",
    
    # identificadores
    "spotify_track_uri", "spotify_episode_uri",
    
    # metadata
    "master_metadata_track_name",
    "master_metadata_album_artist_name",
    "master_metadata_album_album_name",
    "episode_name",
    "episode_show_name",
    
    # consumo
    "ms_played", "minutes_played", "hours_played",
    
    # flags
    "skipped", "is_valid_stream", "is_short_stream", "is_very_short_stream", "is_skipped_custom",
    
    # contexto
    "platform", "conn_country", "reason_start", "reason_end", "shuffle", "offline", "incognito_mode",
    
    # clasificación
    "content_type", "source_type",
    
    # técnico
    "source_file", "offline_timestamp"
]
```
